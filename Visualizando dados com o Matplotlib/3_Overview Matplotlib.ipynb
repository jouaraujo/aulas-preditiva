# Introdução ao Matplotlib e Data Visualization com o Python

O Matplotlib é o principal pacote de visualização de dados do Python. Apesar de ser antigo (foi criado em 2002) ainda tem muita relevância na comunidade e é um pacote essencial no stack de um Cientista de Dados. Ao longo dos anos, o Matplotlib foi ganhando Wrappers (pacotes que usam a API do pacote original, porém são mais fáceis de se trabalhar) muito poderosos. Um deles é o Seaborn que veremos adiante. O próprio Pandas tem um Wrapper de visualização que usa a API do Matplotlib também. Veremos nas próximas aulas como usar suas principais funções e estilos de gráficos para descrever os dados de sua análise.

```python
# Todo gráfico tem ao menos dois componentes, a figura (que contém os elementos do gráfico como texto, 
# cores, tipos de gráficos etc) e os eixos (a caixa que vemos abaixo).

fig = plt.figure()
ax = plt.axes()
```

```python
# Agora que temos a estrutura inicial, vejamos como incluir alguns gráficos.
import numpy as np
valores_de_x = np.arange(10)
valores_de_y = valores_de_x**2

# Perceba que precisamos repetir sempre a mesma estrutura de figura e eixos em cada célula do notebook.
fig = plt.figure()
ax = plt.axes()

# Cria um gráfico simples de linha
ax.plot(valores_de_x,valores_de_y)
```

```python
valores_de_x
```

```python
valores_de_y
```

Se não quisermos escrever as duas linhas de estrutura (figura e eixos) sempre que formos criar os gráficos, <b>podemos usar o comando plt.plot()</b> pois ele cria essas estruturas implicitamente (por debaixo do capô). Veja:

```python
plt.plot(valores_de_x,valores_de_y)
```

Essas facilidades vem da importação de `import matplotlib.pyplot as plt` em vez de `import matplotlib`.

```python
# podemos criar vários gráficos na mesma figura. Veja:

valores_de_x = np.arange(10)
valores_de_y = valores_de_x**2

plt.plot(valores_de_x,valores_de_y)

novos_valores_de_y = 2*valores_de_x+5
plt.plot(valores_de_x,novos_valores_de_y)
```

## Ajustando as cores e os tipos de linhas

```python
# Podemos ajustar as cores com o parâmetro "color"
plt.plot(valores_de_x,valores_de_y, color = 'red')

# Podemos ajustar o tipo de linha com o parâmetro "linestyle"
plt.plot(valores_de_x,novos_valores_de_y,linestyle = '--')
```

```python
# existem outras opções de linhas
plt.plot(valores_de_x,novos_valores_de_y+5,color = 'gray',linestyle = '-') # linha solida
plt.plot(valores_de_x,novos_valores_de_y+8,linestyle = '--') # linha pontilhada 1
plt.plot(valores_de_x,novos_valores_de_y+11,linestyle = '-.') # linha pontilhada 2
plt.plot(valores_de_x,novos_valores_de_y+14,linestyle = ':') # linha pontilhada 3
```

## Inserindo legendas e rótulos

```python
valores_de_x = np.linspace(0,20,100)
valores_de_y = np.cos(valores_de_x) #função cosseno de x
novos_valores_de_y = np.sin(valores_de_x) #função seno de x

# Inserindo legendas para os gráficos (parâmetro "label")
plt.plot(valores_de_x,valores_de_y,label='cosseno(x)') 
plt.plot(valores_de_x,novos_valores_de_y,label='seno(x)')
plt.legend()
```

```python
# Inserindo legendas para os gráficos (parâmetro "label")
plt.plot(valores_de_x,valores_de_y,label='cosseno(x)') 
plt.plot(valores_de_x,novos_valores_de_y,label='seno(x)')

# Insere a legenda fora da figura (mais "clean")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
# Inserindo legendas para os gráficos (parâmetro "label")
plt.plot(valores_de_x,valores_de_y,label='cosseno(x)') 
plt.plot(valores_de_x,novos_valores_de_y,label='seno(x)')

# Insere a legenda fora da figura (mais "clean")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))

# Insere um título nos eixos
plt.xlabel('Valores de x')
plt.ylabel('Valores de y (Funções seno e Cosseno)')

# Insere um título no gráfico
plt.title('Gráficos das funções trigonométricas')
```

# Tipos de gráficos

## Gráficos de Dispersão (Scatter Plot)

```python
valores_de_x = np.linspace(0,20,50)
valores_de_y = np.cos(valores_de_x) #função cosseno de x

plt.plot(valores_de_x,valores_de_y, 'o'); # O parâmetro 'o' significa gráfico de dispersão no método .plot()
```

```python
# Outro exemplo


valores_de_x = np.random.randn(30)
valores_de_y = np.random.randn(30)
cores = np.random.rand(30)

# A função scatter também pode ser usada, porém ela é pouco eficiente para dataframes muito grandes.
plt.scatter(valores_de_x,valores_de_y, c=cores )

plt.colorbar(); #coloca uma barra de cores ao lado do gráfico
```

## Gráficos de Barras (Bar Chart)

```python
valores_de_x = np.arange(10)+1
valores_de_x
```

```python
valores_de_y = valores_de_x*2
valores_de_y
```

```python
plt.bar(valores_de_x,valores_de_y);
```

```python
plt.bar(valores_de_x,valores_de_y,width=0.8);

# adiciona uma nova série de barras no mesmo gráfico (ajustando a largura)
novos_valores_de_x = valores_de_x+0.4
novos_valores_de_y = valores_de_x*4

plt.bar(novos_valores_de_x,novos_valores_de_y,width=0.4);
```

### Barras empilhadas

```python
plt.bar(valores_de_x,valores_de_y,width=0.8,color = 'gray');

# adiciona uma nova série de barras no mesmo gráfico (Barras Empilhadas)
novos_valores_de_x = valores_de_x
novos_valores_de_y = valores_de_x*4

plt.bar(novos_valores_de_x,novos_valores_de_y,width=0.8,bottom=valores_de_y,color='yellow'); # o parâmetro Bottom deve ser usado para incluir a primeira série (a cinza)
```

### Barras horizontais empilhadas

```python
plt.barh(valores_de_x,valores_de_y,height=0.8,color = 'green',label='verde');

# adiciona uma nova série de barras no mesmo gráfico (Barras Horizontais)
novos_valores_de_x = valores_de_x
novos_valores_de_y = valores_de_x*4

plt.barh(novos_valores_de_x,novos_valores_de_y,height=0.8,left=valores_de_y,label='azul');

plt.title('Barras empilhadas horizontais')
plt.legend()
```

## Gráficos de Histograma (Histogram Chart)

Gráficos de histograma são muito úteis para entender a distribuição de uma variável quantitativa. Veja alguns exemplos:

```python
np.random.seed(15)
valores_de_x = np.random.normal(1,5,10000)
```

```python
plt.hist(valores_de_x); # Cria o histograma inicial
```

```python
# podemos mudar a quantidade de Bins (barras) do histograma assim como vários outras opções
plt.hist(valores_de_x,bins=80,color='gray',alpha=0.4);
```

```python
# Exemplo 2: Vários histogramas no mesmo gráfico

np.random.seed(15)
valores_de_x = np.random.normal(1,2,1000)
valores_de_x2 = np.random.normal(6,2,1000)
valores_de_x3 = np.random.normal(10,2,1000)

plt.hist(valores_de_x,bins=15,color='gray',alpha=0.4);
plt.hist(valores_de_x2,bins=15,color='blue',alpha=0.4);
plt.hist(valores_de_x3,bins=15,color='green',alpha=0.4);
```

# Criando uma matriz de gráficos (Subplots)

O Matplotlib é tão flexível que permite inclusive a criação de uma matriz de gráficos na mesma visualização. Vejamos alguns exemplos:

```python
# Tudo começa quando definimos a matriz de gráficos
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2); # Cria o espaço para 4 gráficos
```

<b>Obs:</b> Note que os valores ax1, ax2, ax3 e ax4 são referências para cada gráfico da matriz 2x2.

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2);
ax1.boxplot(df['sepal_length']);
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(2,2);
ax1.boxplot(df['sepal_length']);

valores_de_x = np.random.normal(1,5,100)
ax2.hist(valores_de_x);

valores_de_x = np.random.randint(1,5,20)
valores_de_y = valores_de_x*2
ax3.bar(valores_de_x,valores_de_y,color='green');

valores_de_x = np.linspace(1,5,10)
valores_de_y = np.log(valores_de_x)
ax4.scatter(valores_de_x,valores_de_y,color='red',alpha=0.6);
```

# Visualizando gráficos com o Seaborn

Como dissemos no início das aulas de visualização com o Python, o pacote Matplotlib é muito poderoso, porém exige muito código para conseguir visualizações avançadas. Além disso, o Matplolib não foi concebido para ser usado diretamente com o Pandas. Desta forma, é natural que novos pacotes foram surgindo para melhorar esses pontos. Um deles, muito usado em Ciência de Dados é o Seaborn, que veremos um pouco a seguir. Com o Seaborn, podemos construir visualizações de dados muito bonitas, com pouco código e com mais opções de gráficos pré fabricados muito úteis para a análise de dataframes.

```python
# Importando o Seaborn
import seaborn as sns
```

```python
# Como fizemos em outras ocasiões, vamos importar um dataframe muito conhecido em Ciência de Dados, o "Iris".
df = sns.load_dataset('iris')
df.head()
```

```python
df.dtypes
```

## Gráficos de Dispersão em forma de matriz (Pair plots)

## Gráficos de Histograma em formato de Matriz

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df["time"].value_counts()
```

```python
matriz = sns.FacetGrid(df, row="sex", col="time", margin_titles=True, height=4)
matriz.map(plt.hist, 'tip',bins=10);
```

## Mapas de Calor (Heatmaps)

```python
df = sns.load_dataset('mpg')
df.head()
```

```python
correlacoes = df.corr() # Calcula as correlações de pearson de todas as variáveis quantitativas do dataframe
correlacoes
```

```python
# Cria o mapa de calor do dataframe
sns.heatmap(correlacoes);
```

```python
# Melhorando a aparência do heatmap

plt.figure(figsize=(8,6)) #ajusta o tamanho da figura
sns.heatmap(correlacoes, cmap='RdBu', annot=True); #troca a cor e coloca o valor da correlação dentro da matriz
```
